```python
census = pd.read_csv('/kaggle/input/india-census/india-districts-census-2011.csv')
```

```python
cols = ['District code' , 'District name' , 'Population' , 'Male','Female' , 'Literate' , 'Households_with_Internet']
```

```python
census = census[cols]
```

```python
census.columns
```

```python
final_df = latlong.merge(census ,left_on = 'District' , right_on = 'District name').drop(columns = ['District name'])
```

```python
final_df['Sex_ratio'] = round((final_df['Female']/final_df['Male'])*100)
```

```python
final_df['literacy_rate'] = ((final_df['Literate'] / final_df['Population']) * 100)
```

```python
final_df.drop(columns = ['Male' , 'Female' , 'Literate'] , inplace = True)
```

```python
final_df.to_csv('India.csv', index = False)
```
